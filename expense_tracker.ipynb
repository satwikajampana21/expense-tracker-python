```python
import pandas as pd

data = {
    "Date": [
        "2024-02-01","2024-02-03","2024-02-05",
        "2024-02-10","2024-02-15","2024-02-20",
        "2024-03-01","2024-03-05"
    ],
    "Category": ["Food","Transport","Shopping","Food","Bills","Shopping","Food","Transport"],
    "Amount": [200,100,500,250,800,600,300,150]
}

df = pd.DataFrame(data)
df.to_csv("expenses.csv", index=False)

print("expenses.csv created successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("expenses.csv")
df['Date'] = pd.to_datetime(df['Date'])

print("Expense Data:")
print(df)

# -------------------------------
# 1. Total Expense
# -------------------------------
total_expense = df['Amount'].sum()
print("\nTotal Expense:", total_expense)

# -------------------------------
# 2. Category-wise Expense
# -------------------------------
category_expense = df.groupby('Category')['Amount'].sum()
print("\nCategory-wise Expense:")
print(category_expense)

plt.figure()
category_expense.plot(kind='bar')
plt.title("Category-wise Expense")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()

# -------------------------------
# 3. Monthly Expense Trend
# -------------------------------
df['Month'] = df['Date'].dt.to_period('M')
monthly_expense = df.groupby('Month')['Amount'].sum()
print("\nMonthly Expense Trend:")
print(monthly_expense)

plt.figure()
monthly_expense.plot(kind='line', marker='o')
plt.title("Monthly Expense Trend")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()

# -------------------------------
# Save Results
# -------------------------------
df.to_csv("expense_analysis_results.csv", index=False)
print("\nAnalysis complete. Results saved to expense_analysis_results.csv")
```
